```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
scenarios = ['Blue vs Random Red', 'Blue vs Pretrain Red', 'Blue vs Pretrain Final Red']
average_rewards_red = [-12.017726762823514, 1.1066768374209726, -3.322742873999394]
average_rewards_blue = [-30.46238602143923, -4.757302763204975, -28.330731616564744]

# Tạo vị trí cho các cột
x = np.arange(len(scenarios))  # Số lượng kịch bản
width = 0.35  # Độ rộng mỗi cột

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, average_rewards_red, width, label='Red', color='red', alpha=0.7)
plt.bar(x + width/2, average_rewards_blue, width, label='Blue', color='blue', alpha=0.7)

# Thêm nhãn, tiêu đề, và định dạng
plt.title("Comparison of Average Rewards for Blue and Red Agents", fontsize=16)
plt.xlabel("Scenarios", fontsize=14)
plt.ylabel("Average Reward", fontsize=14)
plt.xticks(x, scenarios, rotation=15)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Hiển thị giá trị trên các cột
for i in range(len(scenarios)):
    plt.text(i - width/2, average_rewards_red[i] + 0.5, f"{average_rewards_red[i]:.2f}", ha='center', fontsize=10, color='black')
    plt.text(i + width/2, average_rewards_blue[i] + 0.5, f"{average_rewards_blue[i]:.2f}", ha='center', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```
